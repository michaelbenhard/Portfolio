```python
import pandas as pd

gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable = '01/31/2020', league_id_nullable = "00")
games = gamefinder.get_data_frames()[0]
#Check what columns are available in the dataframe games
games.columns
#Only take a selected amount of columns
games = games[['TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]
games

games.info()
#Change the data type of 'GAME_DATE' to date time format using pandas function
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])

#Use GAME_DATE as sorting values in increasing order
games = games.sort_values('GAME_DATE')
games

#Calculates the average PLUS_MINUS across the previous 30 games for each team before each game 
#The rolling method's argument closed = 'left', will exclude the current record
games['avg_30_plus)minus'] = games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(30, closed = 'left').mean())

#Shows the 30 first records of the plus minus of the Toronto Raptors
#Note: the NaN value for the Plus Minus is due to the lack of record for the first 30 record, due to the rolling argument of 30
games[games['TEAM_NAME'] == 'Toronto Raptors'].head(35)
```

```python
from pandas.core.generic import Suffixes
gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable = '01/31/2020', league_id_nullable = "00")
games = gamefinder.get_data_frames()[0]
#Check what columns are available in the dataframe games
games.columns
#Only take a selected amount of columns
games = games[['TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]
games['avg_30_plus)minus'] = games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(30, closed = 'left').mean())
games

games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])

#Use GAME_DATE as sorting values in increasing order
games = games.sort_values('GAME_DATE')

games['avg_30_plus_minus'] = games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(30, closed = 'left').mean())
msk = games["MATCHUP"].str.contains("@")

#Separates the dataframe into home and away games
games_away = games[msk]
games_home = games[~msk]

#Combines the two dataframe based on the similar Game_ID and adds label to the home and away team
games_merged = pd.merge(games_home, games_away, on ="GAME_ID", suffixes = ('_home', '_away'))
#Creates a new column for the difference of the 30 games plus between minus away and home
games_merged['avg_30_plus_minus_diff'] = games_merged['avg_30_plus_minus_home'] - games_merged['avg_30_plus_minus_away']

games_merged

#Create a new data frame with only 2 columns and discarding the records with NaN values
games_model =games_merged[['WL_home', 'avg_30_plus_minus_diff']].dropna()
#Changes the WL record into numerical values
games_model['WL_home'] = games_model['WL_home'].map({'W' : 1, 'L': 0})


from sklearn.model_selection import train_test_split

#Splits the dataframe games_model into 80% training and 20% testing set. The argument random_state will ensure reproducibility 
df_train, df_test = train_test_split(games_model, stratify = games_model["WL_home"], test_size =0.2, random_state = 7)

target = "WL_home"
#Create the training and testing data sets and separating them into the features and the data
x_train = df_train.drop(columns = target)
y_train = df_train[target]

x_test = df_test.drop(columns = target)
y_test = df_test[target]

import xgboost as xgb

#Use xgboost classifier
clf = xgb.XGBClassifier(use_label_encoder = False, random_state = 7)
#Fit the training set to the classifier
clf.fit(x_train, y_train)

from sklearn.metrics import accuracy_score
#Predict the testing set using the classifier
y_pred = clf.predict(x_test)
#Check the accuracy of the classifier
accuracy_score(y_test, y_pred)
#How do we improve the accuracy?

#Tuning hyperparameter - random search
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform
#Create a dictionary to figure out the best settings for each parameters
hyp_params={'learning_rate': loguniform(0.0001,1), 
            'max_depth': [2,3,4,5,6,7,8,9], 
            'subsample':[0.7, 0.8,0.9,1.0], 
            'n_estimators': [50,100,150,200]}
random_hyp = RandomizedSearchCV(estimator =clf, param_distributions = hyp_params, n_iter = 20, cv = 7, scoring = 'accuracy', random_state = 7)
random_hyp.fit(x_train, y_train)
#To see the results
random_hyp.best_params_
model_hyp = random_hyp.best_estimator_
y_pred_hyp = model_hyp.predict(x_test)
accuracy_score(y_test, y_pred_hyp)

from joblib import dump, load
dump(model_hyp, 'model_nba.joblib')

model_saved = load('model_nba.joblib')

accuracy_score(y_test, y_pred_hyp)
```

```python
#Given home and away teams, we want to predict which team will win
def predict_games(team_home, team_away):
    gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='01/01/2021',
                                               league_id_nullable='00')
    games = gamefinder.get_data_frames()[0]
    games = games[['TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]
    games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])
    
    msk_home = (games['TEAM_NAME'] == team_home)
    games_30_home = games[msk_home].sort_values('GAME_DATE').tail(30)
    home_plus_minus = games_30_home['PLUS_MINUS'].mean()
    
    msk_away = (games['TEAM_NAME'] == team_away)
    games_30_away = games[msk_away].sort_values('GAME_DATE').tail(30)
    away_plus_minus = games_30_away['PLUS_MINUS'].mean()
    
    games_diff=home_plus_minus - away_plus_minus

    model_saved = load('model_nba.joblib')
    import numpy as np
    #The result of the home team
    predict_home_win =model_saved.predict(np.array([games_diff]))[0]
    #The probability of the previous result happening
    model_saved.predict_proba(np.array([games_diff]))
    #The probability of the home team winning
    predict_winning_probability = model_saved.predict_proba(np.array([games_diff]))[0][1]
    return predict_home_win, predict_winning_probability
```

```python
predict_games("Toronto Raptors", "Boston Celtics")
```
